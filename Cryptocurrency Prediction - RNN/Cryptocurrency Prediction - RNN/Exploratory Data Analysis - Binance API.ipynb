# 1. Importing The Libraries and connecting to the Binance API

```python
from binance1.client import Client
import pandas as pd
from datetime import date
```

```python
apikey = 'YOURAPIKEY'
secret = 'YOURAPISECRET'
```

# 2. Authenticating on Binance

```python
client = Client(apikey, secret)
```

# 3. Get the ticker that we want

```python
depth = client.get_order_book(symbol='ADAUSDT')
depth
```

```python
depth_df = pd.DataFrame(depth['bids'])
depth_df.columns = ['Price', 'Volume']
depth_df.head()
```

# 4. Extracting the data for the crypto that we need

```python
# client.get_historical_klines??
```

```python
historical = client.get_historical_klines('ADAUSDT', Client.KLINE_INTERVAL_1DAY, '1 Jan 2020')
historical
```

```python
hist_df = pd.DataFrame(historical)
hist_df.head(5)
```

# 5. Removing the columns that we don't need and and naming the remaining ones

```python
hist_df.drop(hist_df.columns[7:], axis = 1, inplace=True)
hist_df.head()
```

```python
hist_df.columns = ['Open Time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close Time']
hist_df.head()
```

# 6. Preprocess Historical Data

```python
# Converting the Time units to the normal ones
hist_df['Open Time'] = pd.to_datetime(hist_df['Open Time']/1000, unit='s')
hist_df['Close Time'] = pd.to_datetime(hist_df['Close Time']/1000, unit='s')

hist_df.head()
```

```python
# We can see above and below that we managed to change the 'Open Time' and 'Close Time' formats
```

```python
hist_df.dtypes
```

```python
# Also, our numeric values are represented as strings. We will convert them
```

```python
numeric_columns = ['Open', 'High', 'Low', 'Close', 'Volume']
hist_df[numeric_columns] = hist_df[numeric_columns].apply(pd.to_numeric, axis=1)
hist_df.tail()
```
